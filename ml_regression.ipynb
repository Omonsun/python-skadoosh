```python
imdb_clean=imdbdrop.drop(["id","primaryTitle","originalTitle","isAdult","release_date","directors"],axis=1)
```

```python
from sklearn import  linear_model
from feature_engine.wrappers import SklearnTransformerWrapper
from sklearn.preprocessing import OneHotEncoder
from feature_engine.imputation import CategoricalImputer
from sklearn.impute import KNNImputer
lrn=linear_model.LinearRegression()



imdb_clean_formula=CategoricalImputer(imputation_method="frequent",variables=["genres1","genres2"])

imdb_impute=imdb_clean_formula.fit_transform(imdb_clean)

lrn_feature=SklearnTransformerWrapper(transformer=KNNImputer(n_neighbors=3),variables=["gross"])

imdb_impute2=lrn_feature.fit_transform(imdb_impute)
```

```python
miss.matrix(imdb_impute2)
```

```python
from sklearn.pipeline import Pipeline

from feature_engine.wrappers import SklearnTransformerWrapper
from sklearn.preprocessing import OrdinalEncoder

lrn_encode=SklearnTransformerWrapper(
    transformer=OrdinalEncoder(),
    variables=["genres1","genres2"]
)


imdb_encode=lrn_encode.fit_transform(imdb_impute2)

imdb_encode
####creating data and target array
imdb_data=imdb_encode.drop(["averageRating"],axis=1)
imdb_target=imdb_encode["averageRating"]

```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from feature_engine.selection import SelectBySingleFeaturePerformance
lrn_feature=SklearnTransformerWrapper(
transformer=SequentialFeatureSelector(estimator=KNeighborsRegressor(n_neighbors=5),cv=10,n_features_to_select=4)
)

#lrn_feature=SelectBySingleFeaturePerformance(
 #                   RandomForestRegressor(random_state=42),
  #                  cv=2)

imdb_data_feature=lrn_feature.fit_transform(imdb_data,imdb_target)
```

```python
from sklearn.model_selection import StratifiedKFold,GridSearchCV,RandomizedSearchCV,cross_val_score

ps_lin={
    "fit_intercept": [True,False]
}


#kfold=StratifiedKFold(n_splits=5)

tp=GridSearchCV(estimator=lrn,cv=10,param_grid=ps_lin,scoring="neg_mean_absolute_error",n_jobs=-1)



tp
```

```python
tp.fit(imdb_data,imdb_target).best_params_
```

```python
best_model=tp.fit(imdb_data,imdb_target).best_estimator_
```

```python
best_model
```

```python
cross_val_score(best_model,imdb_data,imdb_target,scoring="neg_mean_absolute_error",cv=10)
```

The closer the score is to 0 (or the more positive it is in the negative form), the better the model's performance
Lower (more negative) scores indicate better performance in terms of mean absolute error
Overall, it seems like our model is performing reasonably well based on the negative mean absolute error scores obtained through cross-validation.
